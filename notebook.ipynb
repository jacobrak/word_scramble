```python
import pandas as pd
df = pd.DataFrame({"text":[]})
text = ""
i = 0
with open("pg1041.txt", "r", encoding="utf-8") as f:
    for line in f:
        if len(line) > 10:
            text += line
        if text:
            df.loc[i] = [text]
            text = ""
            i += 1
```
